```python
#I worked with Brian Shoffeit on this problem
import matplotlib.pyplot as plt
import numpy as np


days = np.array([0, 2, 14, 2, 37, 23, 2, 0, 7, 21, 120, 88, 28, 59, 85, 40, 59, 6])
mean = np.mean(days)
n = len(days)

#Part 1 calculations:
theta = np.arange(.5, 300, .5)
uniform_posterior = (n+1)*(n*mean)**(n+1)/(n*mean+theta)**(n+2)

#Part 2 calculations:
alpha = 1+n
beta = 7+n*mean
gamma_posterior = alpha*beta**alpha/(beta+theta)**(alpha+1)


figure, axis = plt.subplots(figsize = (10, 5))
#Plot for part 1:
uniform_line, = axis.plot(theta, uniform_posterior)
#Plot for part 2:
gamma_line, = axis.plot(theta, gamma_posterior)

axis.set_xlim([0, 300])
axis.set_xlabel("Days between Earthquakes")
axis.set_ylabel("Probability Density")
axis.set_title("Posterior Predictive Distribution of Earthquake Wating Times")
axis.legend([uniform_line, gamma_line], ["Uniform Prior", "Gamma Prior"])
plt.show()
```
