```python
# I worked with Brian Shoffeit on this problem of the homework
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
import math

sample_sizes = [10, 20, 50, 100]
samples = []
val_lambda = 2

# Part a: generates a sample of sizes 10,20,50,100 from the exponential distribution
for i in range(len(sample_sizes)):
    random_generator = np.random.rand(200, sample_sizes[i])
    exponential = np.log(1/random_generator)/val_lambda          #take the inverse of the pdf the exponential distribution
    samples.append(exponential)

# Part b: computes the 200 sample means for each sample size and builds the empirical cdfs
sample_means = []
cdfs = []

for j in range(len(sample_sizes)):
    mean = np.mean(samples[j], axis = 1)
    sample_means.append(mean)
cdfs = [lambda x, i=i: np.size(sample_means[i][sample_means[i] <= x])/200 for i in range (4)]

# Part c: plotting the empirical cdfs against the cdf of the corresponding normal distribution
x_arr = np.arange(0,.9,.001)
norm_cdf = lambda z: (1+erf(z/np.sqrt(2)))/2
figure, axes = plt.subplots(2,2,figsize=(12,9))
axes = axes.ravel()
figure.subplots_adjust(hspace = .3)

for k in range(len(sample_sizes)):
    cdf_plot = np.array([cdfs[k](x) for x in x_arr])
    var = 1/(val_lambda**2*sample_sizes[k])
    z_arr = (x_arr-1/2)/np.sqrt(var)
    normal_cdf_plot = np.array([norm_cdf(z) for z in z_arr])
    axes[k].set_title('Sample Size: '+str(sample_sizes[k]))
    axes[k].plot(x_arr, normal_cdf_plot)
    axes[k].plot(x_arr, cdf_plot)
    axes[k].set_xlim(xmin=.3, xmax=.7)
    axes[k].set_xlabel('x')
    axes[k].set_ylabel('F(x)')

plt.show()
```
